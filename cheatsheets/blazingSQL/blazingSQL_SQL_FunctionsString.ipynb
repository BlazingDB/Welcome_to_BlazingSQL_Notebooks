# SQL String Functions

#### CONCAT

```python
query = '''
    SELECT string
        , 'INFO: ' || string AS r
    FROM df
'''

bc.sql(query)
```

#### SUBSTRING

```python
query = '''
    SELECT string
        , SUBSTRING(string, 0, 2) AS r
    FROM df
'''

bc.sql(query)
```

#### CHAR_LENGTH

```python
query = '''
    SELECT string
        , CHAR_LENGTH(string) AS r
    FROM df
'''

bc.sql(query)
```

#### LEFT

```python
query = '''
    SELECT string
        , LEFT(string, 3) AS r
    FROM df
'''

bc.sql(query)
```

#### RIGHT

```python
query = '''
    SELECT string
        , RIGHT(string, 3) AS r
    FROM df
'''

bc.sql(query)
```

#### LTRIM

```python
query = '''
    SELECT string
        , LTRIM(string) AS r
    FROM df
'''

bc.sql(query)
```

#### RTRIM

```python
query = '''
    SELECT string
        , RTRIM(string) AS r
    FROM df
'''

bc.sql(query)
```

#### REPLACE

```python
query = '''
    SELECT string
        , REPLACE(string, 'RAPIDS', 'NVIDIA RAPIDS') AS r
    FROM df
'''

bc.sql(query)
```

#### UPPER

```python
query = '''
    SELECT string
        , UPPER(string) AS r
    FROM df
'''

bc.sql(query)
```

#### LOWER

```python
query = '''
    SELECT string
        , LOWER(string) AS r
    FROM df
'''

bc.sql(query)
```

#### REVERSE

```python
query = '''
    SELECT string
        , REVERSE(string) AS r
    FROM df
'''

bc.sql(query)
```
