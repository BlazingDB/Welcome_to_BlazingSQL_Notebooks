# SQL Binary Functions

---

#### Addition

```python
query = '''
    SELECT number
        , float_number
        , number + float_number AS r
    FROM df
'''

bc.sql(query)
```

#### Subtraction

```python
query = '''
    SELECT number
        , float_number
        , number - float_number AS r
    FROM df
'''

bc.sql(query)
```

#### Multiplication

```python
query = '''
    SELECT number
        , float_number
        , number * float_number AS r
    FROM df
'''

bc.sql(query)
```

#### Division

```python
query = '''
    SELECT number
        , float_number
        , number / float_number AS r
    FROM df
'''

bc.sql(query)
```

#### MOD

```python
query = '''
    SELECT number
        , float_number
        , MOD(number, 5) AS r
    FROM df
'''

bc.sql(query)
```

#### POWER

```python
query = '''
    SELECT number
        , float_number
        , POWER(number, 2) AS r
    FROM df
'''

bc.sql(query)
```

#### EQUAL TO

```python
query = '''
    SELECT number
        , float_number
        , number = 33 AS r
    FROM df
'''

bc.sql(query)
```

#### NOT EQUAL TO

```python
query = '''
    SELECT number
        , float_number
        , number <> 33 AS r
    FROM df
'''

bc.sql(query)
```

#### LESS THAN

```python
query = '''
    SELECT number
        , float_number
        , number < 32 AS r
    FROM df
'''

bc.sql(query)
```

#### GREATER THAN

```python
query = '''
    SELECT number
        , float_number
        , number > 32 AS r
    FROM df
'''

bc.sql(query)
```

#### LESS THAN OR EQUAL TO

```python
query = '''
    SELECT number
        , float_number
        , number <= 32 AS r
    FROM df
'''

bc.sql(query)
```

#### GREATER THAN OR EQUAL TO

```python
query = '''
    SELECT number
        , float_number
        , number >= 32 AS r
    FROM df
'''

bc.sql(query)
```

#### COALESCE

```python
query = '''
    SELECT word
        , string
        , COALESCE(word, string) AS r
    FROM df
'''

bc.sql(query)
```
