# SQL

#### SELECT

```python
query = '''
    SELECT *
    FROM df
'''

bc.sql(query)
```

```python
query = '''
    SELECT number
        , letter
        , category
    FROM df
'''

bc.sql(query)
```

#### DISTINCT

```python
query = '''
    SELECT DISTINCT category
    FROM df
'''

bc.sql(query)
```

```python
query = '''
    SELECT DISTINCT category
        , letter
    FROM df
'''

bc.sql(query)
```

#### WHERE

```python
query = '''
    SELECT *
    FROM df
    WHERE category = 'B'
'''

bc.sql(query)
```

```python
query = '''
    SELECT *
    FROM df
    WHERE number > 36
'''

bc.sql(query)
```

```python
query = '''
    SELECT *
    FROM df
    WHERE number >= 36
'''

bc.sql(query)
```

#### GROUP BY

```python
query = '''
    SELECT category
        , COUNT(*) AS cnt
    FROM df
    GROUP BY category
'''

bc.sql(query)
```

```python
query = '''
    SELECT category
        , COUNT(*) AS cnt
    FROM df
    GROUP BY category
'''

bc.sql(query)
```

```python
query = '''
    SELECT category
        , letter
        , COUNT(*) AS cnt
    FROM df
    GROUP BY category
        , letter
'''

bc.sql(query)
```

#### HAVING

```python
query = '''
    SELECT category
        , COUNT(*) AS cnt
    FROM df
    GROUP BY category
    HAVING COUNT(*) > 1
'''

bc.sql(query)
```

```python
query = '''
    SELECT category
        , COUNT(*) AS cnt
        , STDDEV(CAST(number AS FLOAT)) AS var
    FROM df
    GROUP BY category
    HAVING STDDEV(CAST(number AS FLOAT)) > 12.0
'''

bc.sql(query)
```

#### ORDER BY

```python
query = '''
    SELECT *
    FROM df
    ORDER BY category
'''

bc.sql(query)
```

```python
query = '''
    SELECT *
    FROM df
    ORDER BY category DESC, number ASC
'''

bc.sql(query)
```

#### CASE

```python
query = '''
    SELECT number
        , CASE WHEN number > 36 THEN 1 ELSE 0 END AS number_cat
    FROM df
'''

bc.sql(query)
```

```python
query = '''
    SELECT CASE WHEN number > 36 THEN 1 ELSE 0 END AS number_cat
        , AVG(float_number) AS float_mean
    FROM df
    GROUP BY CASE WHEN number > 36 THEN 1 ELSE 0 END
    
'''

bc.sql(query)
```

#### LIKE

```python
query = '''
    SELECT string
        , CASE WHEN string LIKE '%GPU%' THEN 1 ELSE 0 END AS Blazing
    FROM df
'''

bc.sql(query)
```

```python
query = '''
    SELECT *
    FROM df
    WHERE string LIKE 'BlazingS__%'
'''

bc.sql(query)
```

#### CAST

```python
query = '''
    SELECT CAST(number AS FLOAT) AS number_float
    FROM df
'''

bc.sql(query)
```

```python
query = '''
    SELECT float_number
        , CAST(float_number AS INTEGER) AS number_float
    FROM df
'''

bc.sql(query)
```

#### LIMIT

```python
query = '''
    SELECT *
    FROM df
    LIMIT 3
'''

bc.sql(query)
```
