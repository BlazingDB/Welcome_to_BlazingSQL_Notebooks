# SQL Unary Functions

#### FLOOR

```python
query = '''
    SELECT float_number
        , FLOOR(float_number) AS f
    FROM df
'''

bc.sql(query)
```

```python
query = '''
    SELECT float_number
        , CEILING(float_number) AS f
    FROM df
'''

bc.sql(query)
```
