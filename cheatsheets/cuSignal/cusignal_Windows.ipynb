# Imports

```python
import cusignal
import cupy as cp
from cupy.fft import fft, fftshift

import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

# cuSignal I/O

Data downloaded from: https://sccn.ucsd.edu/~arno/fam2data/publicly_available_EEG_data.html

# Read sample data

```python
data = cusignal.read_bin(
    '../data/sub-01_ses-1_task-BreathCounting_eeg_SMALL.bdf'
    , dtype=cp.uint8
    , num_samples=200
    , offset=1000
)
```

#### cusignal.windows.windows.barthann()

```python
window = cusignal.barthann(20)
A = fft(window, 2048) / (len(window)/2.0)
freq = cp.linspace(-0.5, 0.5, len(A))
response = 20 * cp.log10(cp.abs(fftshift(A / cp.abs(A).max())))

### Plot
fig, (window_plot, freq_resp) = plt.subplots(2,1, figsize=(9,12))

window_plot.plot(cp.asnumpy(window))
window_plot.set_title("Window")
window_plot.set_ylabel("Amplitude")
window_plot.set_xlabel("Sample")

freq_resp.plot(cp.asnumpy(freq), cp.asnumpy(response))
freq_resp.axis([-0.5, 0.5, -120, 0])
freq_resp.set_title("Frequency response")
freq_resp.set_ylabel("Normalized magnitude [dB]")
freq_resp.set_xlabel("Normalized frequency [cycles per sample]")
```
