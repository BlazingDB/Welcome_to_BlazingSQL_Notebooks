#### cudf.io.json.read_json()

```python
# pandas
import pandas as fmwrk
df = fmwrk.read_csv('../results/pandas_df_with_index.csv')
desc_stats = df.describe()

(
    df
    .groupby(by='category')
    .agg({'num': 'sum', 'char': 'count'})
    .reset_index()
)
```
